####**Univariate non-graphical (UNG)**

#####This analysis will help in identifying distribution, pattern and gain better insight and trend in individual variables

```python
#Group the variable into numeric and categorical

cat_df = df.select_dtypes(include=["object"])
num_df = df.select_dtypes(include=["float64", "int64"])
```

```python
#Create a function for the categorical feature
def cat_univariate_analysis(discrete_data):
  for col_name in discrete_data:
    print(f"Column Name: {col_name}")
    display(discrete_data[col_name].agg(["nunique", "unique" , "count"]))
    display()

    display(discrete_data[col_name].value_counts())
    display()

cat_univariate_analysis(cat_df)
```

```python
#Create function for numeric

def num_univariate_analysis(numerical_data):
  for col_name in numerical_data:
    print(f"Column Name: {col_name}")
    display(numerical_data[col_name].agg(["min", "max", "mean", "median", "std", "count"]))
    display()

num_univariate_analysis(num_df)
```

###**Univariate graphical analysis**

#####This technique uses plots to gain insight into a single variable’s distribution. It provides a quick way of understanding shapes, central tendencies, spreads, modalities, skewnesses, and outliers of our data.

####**Numerical univariate graphical**

```python
#First check the len of column for numerical
num_col = len(num_df.columns)
num_col
```

```python
num_col // 1
```

```python
#Creatng a function to plot the distribution
def univariate_analysis(num_df):
  num_col = len(num_df.columns)
  fig, axes = plt.subplots(nrows=(num_col // 1) + 1, ncols= 3,
                            figsize=(10, 3 * ((num_col // 1)+ 1)))
  fig.tight_layout(pad=5.0)

  axes = axes.flatten()

  for i, col_nam in enumerate(num_df.columns):
    sns.histplot(data = num_df, x= col_nam, kde=True, ax=axes[i])
    #Use kde for smooth distribution
    axes[i].set_title(f"{col_nam}")
    axes[i].set_xlabel(col_nam)
    axes[i].set_ylabel("Frequency")

  #Hide extra space
  for i in range(i + 1, len(axes)):
    fig.delaxes(axes[i])


  plt.show()

univariate_analysis(num_df)
```

#####New_cases and New_death is unimodal and symmetry.
#####The distribution of 'new_cases' appears to be positively skewed. This means that the tail of the distribution extends more to the right (higher values).The distribution suggests that on most days, the number of new cases was relatively low to moderate.However, there were periods with much higher surges of infections, leading to the positive skew hence the unimodal distribution

#####The distribution of 'new_death' also appears to be positively skewed, although perhaps less pronounced than 'new_cases'. The tail extends to the right, indicating some days with a higher number of deaths. The right tail indicates that while most days had a lower number of deaths, there were unfortunately some days with substantially higher fatalities. The positive skew reflects these less frequent but more severe days.

#####**Daily Trends in COVID-19 Cases and Deaths Over Time**

```python
# Ensure the 'date' column is in datetime format
df['date'] = pd.to_datetime(df['date'])

# Sort the dataframe by date to ensure the line plot is correct
df.sort_values(by='date', inplace=True)

# Plotting the daily trends
plt.figure(figsize=(13, 8)) # Adjust figure size as needed

sns.lineplot(data=df, x='date', y='new_cases', label='Daily New Cases')
sns.lineplot(data=df, x='date', y='new_death', label='Daily New Deaths')

plt.title('Daily Trends in COVID-19 Cases and Deaths Over Time')
plt.xlabel('DATE')
plt.ylabel('COUNT')
plt.legend()
plt.grid(True) # Add a grid for easier reading
# Set the x-axis ticks to show every date
plt.xticks(df['date'], rotation=90)# Rotate x-axis labels for better readability
plt.tight_layout() # Adjust layout to prevent labels overlapping

plt.show()
```

#####**Average Daily Trends of Cases, Deaths, and Vaccinations by Day of the Week**

```python
# Extract the day of the week
df['day_of_week'] = df['date'].dt.day_name()

# Group by day of the week and calculate the mean for each metric
daily_trends = df.groupby('day_of_week')[['new_cases', 'new_death', 'vaccinated']].mean().reset_index()

# Define the order of the days of the week
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Reindex the dataframe to ensure the correct order of days
daily_trends['day_of_week'] = pd.Categorical(daily_trends['day_of_week'], categories=day_order, ordered=True)
daily_trends = daily_trends.sort_values('day_of_week')

# Melt the DataFrame for easier plotting
daily_trends_melted = daily_trends.melt(id_vars='day_of_week', var_name='Metric', value_name='Average Count')

# Create the line plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=daily_trends_melted, x='day_of_week', y='Average Count', hue='Metric', marker='o')

plt.title('Average Daily Trends of Cases, Deaths, and Vaccinations by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Average Count')
plt.grid(True)
plt.show()
```

###**Cumulative trends in infections and fatalities**
#####**Cumulative Trends in COVID-19 Cases and Deaths Over Time**

```python
# Calculate cumulative cases and deaths
df['cumulative_cases'] = df['new_cases'].cumsum()
df['cumulative_death'] = df['new_death'].cumsum()

# Plotting the cumulative trends
plt.figure(figsize=(12, 6))

sns.lineplot(data=df, x='date', y='cumulative_cases', label='Cumulative New Cases')
sns.lineplot(data=df, x='date', y='cumulative_death', label='Cumulative New Deaths')

plt.title('Cumulative Trends in COVID-19 Cases and Deaths Over Time')
plt.xlabel('Date')
plt.ylabel('Cumulative Count')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

###**Insight:**
1. **Steady Growth in Cumulative Cases:** The cumulative new cases show a relatively consistent upward trend throughout the month, although the rate of increase appears to slightly accelerate in the latter half of November. This signifies a continuous accumulation of new infections.
2. **Slow Growth in Cumulative Deaths:** The cumulative new deaths show a very gradual upward trend, indicating that while fatalities were occurring, the rate at which the total number of new deaths accumulated was significantly slower than the accumulation of new cases.
3. **Divergence Over Time:** The gap between the cumulative new cases and cumulative new deaths widens considerably as the month progresses. This further emphasizes the much higher volume of new infections compared to new deaths during this period.

###**Patterns:**

1. **Consistent Accumulation:** The primary pattern is the steady and continuous accumulation of both new cases and new deaths over time, as expected with cumulative data.
2. **Disproportionate Scale:** The most striking pattern is the vast difference in scale between the cumulative new cases and cumulative new deaths, highlighting a lower mortality rate relative to the number of new infections within this timeframe.
